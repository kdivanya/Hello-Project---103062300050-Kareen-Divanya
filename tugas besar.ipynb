```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
n = 100
X = np.random.rand(n)
Y = 2 * X + np.random.rand(n)

plt.scatter(X,Y)
plt.xlabel('Variabel Independen (X)')
plt.ylabel('Variabel Dependen (Y)')
plt.title('Hubungan antara X dan Y')
plt.show()

correlation = np.corrcoef(X, Y)[0, 1]
print("Korelasi Pearson antara X dan Y:", correlation)

coefficients = np.polyfit(X,Y,1)
trend = np.poly1d(coefficients)
plt.plot(X, Y, 'o', X, trend(X), '-')
plt.xlabel('Variabel Independen (X)')
plt.ylabel('Variabel Independen (Y)')
plt.title('Tren Data X dan Y')
plt.show()
```

**Penjelasan Setiap Baris:**

1. import numpy as np untuk mengimpor library numpy dengan alias np untuk digunakan dalam pembuatan data dummy
2. import matplotlib.pyplot as plt untuk mengimpor library matplotlib.plyplot dengan alias plt untuk menampilkan grafik.
3. np.random.seed(0) menetapkan seed untuk memastikan hasil acak yang konsisten setiap kali program dijalankan
4. n = 100 untuk menetapkan jumlah sampel data
5. X = np.random.rand(n) untuk membuat variabel independen(X) dengan nilai acak dari distribusi uniform.
6. Y = 2 * X + np.random.rand(n) untuk membuat variabel dependen(Y) dengan hubungan linear terhadap X dan ditambah dengan noise acak dari distribusi normal.
7. plt.scatter(X, Y) untuk menampilkan scatter plot dari data X dan Y.
8. plt.xlabel, plt.ylabel, plt.title untuk memberi judul plot dan label sumbu.
9. plt.show() untuk menampilkan plot
10. correlation = np.corrcoef(X,Y)[0,1] untuk menghitung korelasi  pearson antara X dan Y.
11. print("Korelasi Pearson antara X dan Y:", correlation) untuk menampilkan nilai korelasi pearson.
12. coefficients = np.polyfit (X,Y,1) untuk menghitung koefisien regresi linear antara X dan Y.
13. trend = np.poly1d(coefficients) untuk membuat fungsi polinomial berdasarkan koefisien yang sudah dihitung.
14. plt.plot(X,Y,'o',X,trend(X),'-') untuk menampilkan plot dari data dan trendline.

**Hasil Eksplorasi Data:**

- Nilai korelasi pearson menunjukkan seberapa kuat hubungan linear antara X dan Y. Nilai ini berkisar dari -1 (hubungan negatif sempurna) hingga 1. Hubungan linear menjadi lebih kuat semakin dekat dengan 1. Tidak ada hubungan linear yang signifikan antara kedua variabel jika nilainya mendekati 0.
- Garis tren menunjukkan tren data. Dalam hal ini, garis tren akan mendekati garis lurus karena hubungan antara X dan Y adalah linier (Y = 2X + noise).

```python
import random
import matplotlib.pyplot as plt

random_data = [random.randint(10, 50) for _ in range(150 - 14)]
data = [10, 20, 20, 30, 10, 20, 30, 40, 10, 30, 40, 50, 40, 30] + random_data

data_freq = {}
for item in data:
  if item in data_freq:
    data_freq[item] += 1
  else:
    data_freq[item] = 1

print("Data\tFrekuensi")
for key, value in data_freq.items():
  print(f"{key}\t{value}")

plt.bar(data_freq.keys(), data_freq.values())
plt.xlabel('Data')
plt.ylabel('Frekuensi')
plt.title('Grafik Data dan Frekuensi')
plt.show()
```

```python
import numpy as np
from scipy.stats import t

penjualan_elektronik = [10, 9, 12, 6, 8, 12, 7, 6, 3, 7, 7, 6, 6, 3, 7, 5, 7, 3, 7, 5, 5, 10, 7, 5]
penjualan_komputer = [7, 5, 4, 11, 7, 4, 5, 9, 3, 11, 4, 5, 6, 6, 6, 5, 6, 5, 7, 4, 6, 4, 6, 9]
penjualan_aksesoris = [52, 50, 12, 6, 6, 7, 11, 57, 48, 60, 60, 55, 54, 59, 60, 55, 54, 56, 57, 54, 58, 60, 57, 53]
penjualan_health_beauty = [5, 8, 7, 6, 9, 4, 7, 6, 8, 7, 5, 6, 9, 7, 5, 8, 6, 7, 6, 8, 4, 5, 4, 6]

mean_elektronik = np.mean(penjualan_elektronik)
std_dev_elektronik = np.std(penjualan_elektronik, ddof=1)
n_elektronik = len(penjualan_elektronik)

mean_komputer = np.mean(penjualan_komputer)
std_dev_komputer = np.std(penjualan_komputer, ddof=1)
n_komputer = len(penjualan_komputer)

mean_aksesoris = np.mean(penjualan_aksesoris)
std_dev_aksesoris = np.std(penjualan_aksesoris, ddof=1)
n_aksesoris = len(penjualan_aksesoris)

mean_health_beauty = np.mean(penjualan_health_beauty)
std_dev_health_beauty = np.std(penjualan_health_beauty, ddof=1)
n_health_beauty = len(penjualan_health_beauty)

t_score = t.ppf(0.975, df=n_elektronik - 1)

CI_elektronik = t_score * (std_dev_elektronik / np.sqrt(n_elektronik))
CI_komputer = t_score * (std_dev_komputer / np.sqrt(n_komputer))
CI_aksesoris = t_score * (std_dev_aksesoris / np.sqrt(n_aksesoris))
CI_health_beauty = t_score * (std_dev_health_beauty / np.sqrt(n_health_beauty))

print("Estimasi Rata-rata Penjualan Bulanan per Kategori Produk (Interval Kepercayaan 95%):")
print("Elektronik:", round(mean_elektronik, 2), "+/-", round(CI_elektronik, 2))
print("Komputer:", round(mean_komputer, 2), "+/-", round(CI_komputer, 2))
print("Aksesoris:", round(mean_aksesoris, 2), "+/-", round(CI_aksesoris, 2))
print("Health & Beauty:", round(mean_health_beauty, 2), "+/-", round(CI_health_beauty, 2))
```
